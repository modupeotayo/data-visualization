## How to create a graph in Python?
One of the most commonly used libraries in Python for data visualization is `matplotlib`. There are also other common graphing libraries that we will explore in the next lessons.
I invite you to download the library, for example by typing the following command in your terminal: `pip install matplotlib`
We can now import the library into our notebook.

```python
import matplotlib.pyplot as plt
```

### Our first graph
Making a graph is very simple:

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

However our graph is incomplete and missing crucial elements: 
* a title
* a label for the x-axis
* a label for the y-axis

It's very simple to add these elements to your graph:

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("Growth curve of a colony of bacteria")
plt.xlabel("hours")
plt.ylabel("number of bacteria")
plt.show()
```

By convention, data that is independent of our observations will be on the x-axis (in this case, time), and the data that we observe will be on the y-axis (in this case, the number of bacteria). 

We can also represent multiple curves on the same graph. To distinguish them, we will use legends.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label="bacteria A")
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label="bacteria B")

plt.title("Growth curve of two colonies of bacteria")
plt.xlabel("hours")
plt.ylabel("number of bacteria")
plt.legend()
plt.show()
```

## Different types of graphs
As we've seen previously, there are different types of data (quantitative and qualitative). To represent this diversity of data, there is also a variety of possible visualizations and graphs. All of these graphs are more or less suitable depending on the context in which you use them.

### Line graphs
Line graphs are the ideal visualization when you want to show:

* The trend or evolution of an element.
* The evolution of multiple elements in relation to each other.
* Highlight anomalies within and between data series.

When time is one of your data points, you can use it as an indicator to consider using a line graph.

The type of data used in line graphs consists of numerical and continuous data on both axes. Categorical data cannot be displayed on a line graph.

You can create a line graph using the `plt.plot()` function with Matplotlib

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("Growth curve of a colony of bacteria")
plt.xlabel("hours")
plt.ylabel("number of bacteria")
plt.show()
```

Documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html  
Examples: https://www.python-graph-gallery.com/line-chart/

### Bar graphs
A bar graph is a common type of chart for comparing categorical data. It consists of several rectangles aligned on a common baseline. The height of each rectangle is proportional to the value it represents. Bar graphs are often used to answer the question 'How much?' for each category.

This type of graph is very easy to read because our eyes are adept at comparing the heights of objects side by side. We can quickly identify the smallest and largest bars, as well as the differences in height between the bars.

The types of data used in a bar graph are categorical data on the x-axis and numerical and discrete data on the y-axis.

You can create a bar graph using the `plt.bar()` function with Matplotlib.

```python
plt.bar(['banana', 'apple', 'pear', 'orange'], [3.34, 3.0, 3.55, 3.25])
plt.title("Prix of fruits")
plt.xlabel("fruits")
plt.ylabel("price per kg")
plt.show()
```

### Histogram 

An histogram takes as input a numerical variable only. The variable is divided into several 'bins,' and the number of observations per bin is represented by the height of the bar. An histogram is primarily used to examine the overall distribution of a variable.

Unlike bar graphs, here the width of the bars is significant and conveys information. The more bins you have, the more bars you will have, and they will be narrower, providing more detail in your distribution. The fewer bins you have, the fewer bars there will be, and they will be wider, giving you less detail in your distribution.

There is no consensus on the appropriate number of bins to use, so I encourage you to experiment with different numbers of bins to see the different information you can extract.

You can create a histogram using the `plt.hist()` function with Matplotlib.

```python
students_age = [ 18, 20, 20, 20, 21, 21,21, 21, 22, 22, 22, 22, 22, 23, 23, 23 ,24, 24, 24, 24, 25, 25, 26, 27, 31]
plt.hist(students_age, bins = 6)
plt.title("Distribution of students' ages.")
plt.xlabel("age of students")
plt.ylabel("number of students")
plt.show()
```

Documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
Exemples: https://www.python-graph-gallery.com/histogram/

### Pie charts 
Pie charts are preferably used when you are trying to compare the distribution of parts within a whole. This type of chart represents numbers in percentages, and the sum of all the parts of the pie should be equal to 100%. When you have 'percentages of...' or 'parts of...', you can use that as an indicator to consider using a pie chart.

Pie charts are widespread but unfortunately often misused. We will see why in our next class.

We will, therefore, keep the following two main use cases:

If you want your audience to have a general idea of the relationship between parts in your data, and the precise size comparison of the slices is less important.
To highlight a very small or very large part in your data.
The type of data used in a pie chart is categorical data.

You can create a pie chart using the `plt.pie()` function with Matplotlib.

```python
plt.pie([0.6, 0.3, 0.1], labels=['Product A', 'Product B', 'Product C'])
plt.title("Distribution of products sold this year")
plt.show()
```

### Box plot
A box plot is a graph that presents multiple statistical measures to help us understand the distribution of a variable. The type of data used here is numerical data.

Here is a description of the statistical measures that can be observed on a box plot:

* Range: The difference between the highest and lowest values in the data set.
* Percentiles: A measure that indicates the value below which a given percentage of observations falls in a group of observations.
* Quartiles: Values that divide the data points into four more or less equal parts, or quartiles.
* Interquartile Range (IQR): A measure of statistical dispersion and variability based on dividing a data set into quartiles. IQR = Q3 - Q1.
* Variance: The average squared difference of values from the mean to measure the dispersion of a data set from the mean.
* Standard Deviation: The standard difference between each data point and the mean, and the square root of the variance.
* Outlier: An outlier is an observation that is unusually far from other values in a random sample from a population. We consider a value an outlier if it is smaller than Q1 - 1.5 IQR or larger than Q3 + 1.5 IQR.

We can see how all these metrics are summarised in a box plot: https://byjus.com/maths/box-plot/

We can create a box plot using the `plt.boxplot()` function with Matplotlib.

```python
plt.boxplot(students_age)
plt.title("Distribution of students ages")
plt.xlabel("students")
plt.ylabel("age")
plt.show()
```

Documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

Examples: https://www.python-graph-gallery.com/boxplot/

### Scatter plot
A scatter plot is a type of diagram that illustrates the relationship between two numerical variables. Each measurement in the dataset is represented with a point in this diagram. Scatter plots are very useful when you want to understand the distribution of your data, observe a trend between your variables, or identify outliers.

We can identify a linear trendline using the Ordinary Least Squares method and correlation. This method tries to find the line closest to all the points in our dataset. If this line is increasing, it means that our variables are increasing together, and the correlation is positive. If this line is decreasing, it means that when one variable increases, the other variable decreases, and the correlation is negative. It is also possible to find no relationship between the two variables. This is the case if the line is nearly constant, and the correlation is around 0.

To experiment with these concepts: https://seeing-theory.brown.edu/regression-analysis/index.html#section1

We can create a scatter plot using the `plt.scatter()` function with Matplotlib

```python
students_height = [170, 165, 172, 192, 185, 167, 184, 189, 156, 166, 178, 170, 172, 185, 184, 176, 197, 168, 169, 181, 183, 176, 177, 155, 189]
plt.scatter(students_height, students_age)
plt.title('Distribution of students height and students age')
plt.xlabel("height")
plt.ylabel("age")
plt.plot()
```

## Matplotlib and pandas
You may have noticed that in each of the previous code cells, the data to plot the graph was provided in the form of lists. However, as you have seen in previous lessons, we mainly work with DataFrames using the pandas library.

Fortunately, thanks to dedicated developers, we can use matplotlib with pandas directly. We can use the plot() method directly. This pandas approach has the advantage of having a very lightweight syntax. However, the DataFrame must be perfectly formatted for correct rendering.

```python
import pandas as pd

#Let's create a dataframe
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [10, 25, 15, 30]}
df = pd.DataFrame(data)

# Create a bar graph
df.plot.bar(x='Category', y='Value')
```

Alternatively, we can simply pass a DataFrame as a parameter to the matplotlib functions. This provides much more flexibility and control over the graph's rendering than the first method so we will prefer using this method.

```python
plt.bar(df['Category'], df['Value'])
plt.title("Values for different categories")
plt.ylabel("Values")
plt.show()
```

## Object-oriented Matplotlib vs. Matplotlib pyplot


There are two ways to use Matplotlib:

* The pyplot interface
* The object-oriented interface

The examples we have seen in this course so far rely on the pyplot interface. Now, we will see how to use Matplotlib with the object-oriented interface. This interface has the advantage of giving us more flexibility for composing our graphs. It is also preferred when dealing with multiple graphs and references to graphs that need to be passed into functions or dashboard.

In general, try to use the object-oriented interface rather than the pyplot interface. With the OOP interface, we start by creating an instance of a Figure, which provides an empty canvas. The Figure object is the one that will contain all the other elements that make up the graphical image.

```python
import matplotlib.pyplot as plt
fig = plt.figure()
```

Now, we need to add Axes to the figure. The add_axes() method takes an argument, which is a list of 4 elements corresponding to the left, bottom, width, and height of the figure. Each number should be between 0 and 1.

Be careful not to confuse 'Axes,' which is a reference to an individual plot within a figure, and 'Axis,' which refers to the x and y of a plot.

```python
ax = fig.add_axes([0,0,1,1])
```

We can also add legends to our graphs using the following methods of the Axes class:

set_title(), to add a title.
set_x_label(), to add a label to our x-axis.
set_y_label(), to add a label to our y-axis.

```python
ax.set_title("Growth curve of a colony of bacteria")
ax.set_xlabel("hours")
ax.set_ylabel("number of bacteria")
```

Next, we can draw our graph using the following method of the Axes class: plot()

```python
ax.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

Finally, we can display our graph.

```python
fig
```

Similarly, the functions for different types of graphs have the same names in the object-oriented interface as in the pyplot interface:

- [`bar()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html)
- [`hist()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html)
- [`pie()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html)
- [`box()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.boxplot.html)
- [`scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html)
